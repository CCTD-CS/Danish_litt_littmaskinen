```python
# set directories

input_dir = "raw_files/pop/DK_ForSidsteAkt.txt"
output_dir = "processed_files/pop/DK_ForSidsteAkt.txt"


temp_dir = "temp/temp.txt"
```

```python
# Chop into sentences

import spacy

# Load Danish spaCy model
nlp = spacy.load("da_core_news_sm")

# Read the text file
with open(input_dir , "r", encoding="utf-8") as file:
    text = file.read().replace('\f', '')

# Process the text
doc = nlp(text)

# Extract sentences
sentences = [sent.text.strip() for sent in doc.sents]

# Example: print first 5 sentences
for sent in sentences[:5]:
    print(sent)
```

```python
# Save sentences to a new file
with open(temp_dir, "w", encoding="utf-8") as output_file:
    for sentence in sentences:
        output_file.write(sentence + "\n") 
```

```python
# Remove empty lines

with open(temp_dir, "r", encoding="utf-8") as infile, \
     open(output_dir, "w", encoding="utf-8") as outfile:
    for line in infile:
        if (line.strip()):  # Only write non-empty, non-whitespace lines
            outfile.write(line.strip() + "\n")
```

```python
# function that sentenizes the text fiels

def sentenize_text_file(input_dir, output_dir):
    # Load Danish spaCy model
    nlp = spacy.load("da_core_news_sm")

    # Read the text file
    with open(input_dir , "r", encoding="utf-8") as file:
        text = file.read().replace('\f', '')

    # Process the text
    doc = nlp(text)

    # Extract sentences
    sentences = [sent.text.strip() for sent in doc.sents]

    # Example: print first 5 sentences
    for sent in sentences[:5]:
        print(sent)

    # Save sentences to a new file
    with open("temp/temp.txt", "w", encoding="utf-8") as output_file:
        for sentence in sentences:
            output_file.write(sentence + "\n")

    # Remove empty lines
    with open("temp/temp.txt", "r", encoding="utf-8") as infile, \
        open(output_dir, "w", encoding="utf-8") as outfile:
        spillover = ""
        for line in infile:
          if (line.strip()):  # Only write non-empty, non-whitespace lines
                if len(line.strip()) == 1:
                    spillover = line.strip()
                else:
                    outfile.write(spillover + line.strip() + "\n")
                    spillover = ""
```

```python
 # pop
sentenize_text_file("raw_files/pop/DK_ForSidsteAkt.txt", "processed_files/pop/DK_ForSidsteAkt.txt")
sentenize_text_file("raw_files/pop/DK_Isola.txt", "processed_files/pop/DK_Isola.txt")
sentenize_text_file("raw_files/pop/DK_MysterietIGenbrugsen.txt", "processed_files/pop/DK_MysterietIGenbrugsen.txt")
sentenize_text_file("raw_files/pop/DK_PigenUnderTraeet.txt", "processed_files/pop/DK_PigenUnderTraeet.txt")
sentenize_text_file("raw_files/pop/DK_RigeSvin.txt", "processed_files/pop/DK_RigeSvin.txt")

# world
sentenize_text_file("raw_files/world/DK_AlleHimlensFugle.txt", "processed_files/world/DK_AlleHimlensFugle.txt")
sentenize_text_file("raw_files/world/DK_Dafnesyndromet.txt", "processed_files/world/DK_Dafnesyndromet.txt")
sentenize_text_file("raw_files/world/DK_HummerensSkjold.txt", "processed_files/world/DK_HummerensSkjold.txt")
sentenize_text_file("raw_files/world/DK_LogTraekkerTarer.txt", "processed_files/world/DK_LogTraekkerTarer.txt")
sentenize_text_file("raw_files/world/DK_PengePaLommen.txt", "processed_files/world/DK_PengePaLommen.txt")

```
