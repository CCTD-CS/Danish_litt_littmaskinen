```python
import spacy

# Load Danish spaCy model
nlp = spacy.load("da_core_news_sm")

# Read the text file
with open("DK_DenTavskeEnke.txt", "r", encoding="utf-8") as file:
    text = file.read().replace('\f', '')

# Process the text
doc = nlp(text)

# Extract sentences
sentences = [sent.text.strip() for sent in doc.sents]

# Example: print first 5 sentences
for sent in sentences[:5]:
    print(sent)

```

```python
# Save snetences to a new file
with open("sentences_DK_PengePaLommen.txt", "w", encoding="utf-8") as output_file:
    for sentence in sentences:
        output_file.write(sentence + "\n") 
```

```python
# Remove empty lines from the output file
input_path = "sentences_DK_DenTavskeEnke.txt"
output_path = "sentences_NoEmptyLines_DK_DenTavskeEnke.txt"

with open(input_path, "r", encoding="utf-8") as infile, \
     open(output_path, "w", encoding="utf-8") as outfile:
    for line in infile:
        if line.strip() and line.strip() != "-":  # Only write non-empty, non-whitespace lines
            outfile.write(line.strip() + "\n")
```
